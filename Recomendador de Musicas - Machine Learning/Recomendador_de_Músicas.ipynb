Acústica: Medida de confiança de 0,0 a 1,0 se a faixa é acústica.

Dançabilidade: Descreve o quão adequada uma faixa é para dançar com base em uma combinação de elementos musicais, incluindo tempo, estabilidade do ritmo, força da batida e regularidade geral. Um valor de 0,0 é o menos dançável e 1,0 é o mais dançável.

Duração_ms: Duração da trilha em milissegundos.

Energia: Variável numérica, Energia é uma medida de 0,0 a 1,0 e representa uma medida perceptiva de intensidade e atividade. Normalmente, as faixas energéticas parecem rápidas, altas e barulhentas. Por exemplo, o death metal tem alta energia, enquanto um prelúdio de Bach tem uma pontuação baixa na escala. As características perceptivas que contribuem para este atributo incluem faixa dinâmica, intensidade percebida, timbre, taxa de início e entropia geral.

Explícito: Variável categórica, se a faixa tem ou não letras explícitas (verdadeiro = sim (1); falso = não(0), não OU desconhecido).

Instrumentalidade: Prevê se uma faixa não contém vocais. Os sons “Ooh” e “aah” são tratados como instrumentais neste contexto. Faixas de rap ou de palavras faladas são claramente “vocais”. Quanto mais próximo o valor de instrumentalidade estiver de 1,0, maior a probabilidade de a faixa não conter conteúdo vocal. Valores acima de 0,5 destinam-se a representar faixas instrumentais, mas a confiança é maior à medida que o valor se aproxima de 1,0.

Tonalidade: A tonalidade geral estimada da faixa. Os inteiros são mapeados para pitchs usando a notação padrão de Pitch Class. Por exemplo. 0 = C, 1 = C#/Db, 2 = D, e assim por diante. Se nenhuma tonalidade foi detectado, o valor é -1.

Ao vivo: Detecta a presença de um público na gravação. Valores mais altos de vivacidade representam uma probabilidade maior de que a faixa tenha sido executada ao vivo. Um valor acima de 0,8 fornece uma forte probabilidade de que a faixa esteja ativa.

Volume em dB: Volume geral de uma faixa em decibéis (dB). Os valores de volume são calculados em média em toda a faixa e são úteis para comparar o volume relativo das faixas. A sonoridade é a qualidade de um som que é o principal correlato psicológico da força física (amplitude). Os valores típicos variam entre -60 e 0 db.

Modalidade: Indica a modalidade (maior ou menor) de uma faixa, o tipo de escala da qual seu conteúdo melódico é derivado. Maior é representado por 1 e menor é 0.

Popularidade: A popularidade de uma faixa é um valor entre 0 e 100, sendo 100 o mais popular. A popularidade é calculada por algoritmo e é baseada, em grande parte, no número total de execuções que a faixa teve e quão recentes são essas execuções.

Fala: A fala detecta a presença de palavras faladas em uma faixa. Quanto mais exclusivamente falada a gravação (por exemplo, talk show, audiolivro, poesia), mais próximo de 1,0 o valor do atributo. Valores acima de 0,66 descrevem faixas que provavelmente são feitas inteiramente de palavras faladas. Valores entre 0,33 e 0,66 descrevem faixas que podem conter música e fala, seja em seções ou em camadas, incluindo casos como música rap. Os valores abaixo de 0,33 provavelmente representam músicas e outras faixas que não são de fala.

Ritmo: Variável numérica, Tempo estimado geral de uma faixa em batidas por minuto (BPM). Na terminologia musical, tempo é a velocidade ou ritmo de uma determinada peça e deriva diretamente da duração média da batida.

Valência: Variável numérica, Medida de 0,0 a 1,0 descrevendo a positividade musical transmitida por uma faixa. Faixas com alta valência soam mais positivas (por exemplo, feliz, alegre, eufórica), enquanto faixas com baixa valência soam mais negativas (por exemplo, triste, deprimida, irritada).

##Preparação dos Dados

```python
dados.Artista.value_counts()
```

```python
artist_counts = dados["Artista"].value_counts()
dados["Artista"] = dados["Artista"].apply(lambda x: 'Outros' if artist_counts[x] == 1 else x)
```

```python
dados.Artista.value_counts()
```

```python
ohe = OneHotEncoder(dtype=int)
colunas_ohe = ohe.fit_transform(dados[["Artista"]]).toarray()
dados2 = dados.drop("Artista", axis = 1)

n_dados = pd.concat([dados2, pd.DataFrame(colunas_ohe, columns=ohe.get_feature_names_out(["Artista"]))], axis=1)
```

```python
n_dados = n_dados.drop(["Nome", "Música - Artista", "id"], axis=1)
```

```python
n_dados.head(1)
```

```python
n_dados.shape
```

##Redução de Dimensionalidade

```python
SEED = 467
np.random.seed(467)

tarefa_pipeline = Pipeline([("scaler", StandardScaler()), ("PCA", PCA(n_components=0.7, random_state=SEED))])

modelo = tarefa_pipeline.fit_transform(n_dados)
projeto = pd.DataFrame(data=modelo)
```

```python
tarefa_pipeline[1].n_components_
```

##K-Means

```python
kmeans_pca = KMeans(n_clusters=75, verbose=False, random_state=SEED)
kmeans_pca.fit(projeto)

dados["cluster"] = kmeans_pca.predict(projeto)
projeto["cluster"] = kmeans_pca.predict(projeto)
projeto["música"] = dados["Música - Artista"]

```

##Plotando a Cluster

```python
fig = px.scatter(
    projeto, x=0, y=1, color="cluster", hover_data=[0, 1, "música"])
fig.show()
```

##Sistema de Recomendação

```python
musica_escolhida = "Paradise - Coldplay"
```

```python
cluster = list(projeto[projeto["música"]== musica_escolhida]["cluster"])[0]
musicas_recomendadas = projeto[projeto["cluster"]== cluster][[0, 1, "música"]]
x_musica = list(projeto[projeto["música"]== musica_escolhida][0])[0]
y_musica = list(projeto[projeto["música"]== musica_escolhida][1])[0]

distancias = euclidean_distances(musicas_recomendadas[[0, 1]], [[x_musica, y_musica]])
musicas_recomendadas["distancias"]= distancias
recomendada = musicas_recomendadas.sort_values("distancias").head(2)
recomendada
```
